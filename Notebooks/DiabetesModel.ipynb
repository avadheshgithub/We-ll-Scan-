```python
# how to load dataset of csv file
import pandas as pd
```

```python
df = pd.read_csv('Diebetesdata.csv')
```

```python
X = df.iloc[:,0:11]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 2)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Create Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
dieb_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
#  Train the Model
dieb_classifier.fit(X_train, y_train)
```

```python
#  Evaluate Model Accuracy
y_pred1 = dieb_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred1)
print("Accuracy:", accuracy)
```
